# Data Preprocessing/ Cleaning

## Make new dataframe by dropping the column which aren't useful

```python
business = df.drop(['Country', 'Postal Code'], axis=1)
```

```python
business.head()
```

# Relationship Analysis

```python
# Corelation Matrix
corelation = business.corr()
```

## Heatmap

```python
sns.heatmap(corelation, xticklabels=corelation.columns, yticklabels=corelation.columns,
            annot=True)
```

## Pairplot

```python
sns.pairplot(business) # pairwise relationship in whole dataset
```

## Relational Scatter Plot

```python
sns.relplot(x='Sales', y='Profit', hue='Ship Mode', data=business)
```

```python
sns.relplot(x='Sales', y='Profit', hue='Category', data=business)
```

```python
sns.relplot(x='Sales', y='Discount', data=business)
```

## Distribution Plot

```python
sns.distplot(business['Profit'])
```

## Categorical Plot

```python
sns.catplot(x="Profit", kind='box', data=business)
```

## Bar plot

```python
plt.figure(figsize= (10,16))
business.groupby('Category')['Profit','Sales'].agg(['sum']).plot.bar()
plt.ylabel('Profit')
plt.show()
```

```python
plt.figure(figsize= (10,16))
business.groupby('Region')['Profit','Region'].agg(['sum']).plot.bar()
plt.ylabel('Profit')
plt.show()
```

```python
# For Sub-categories
plt.figure(figsize=(14,12))
statewise = business.groupby(['Sub-Category'])['Profit'].sum().nlargest(50)
statewise.plot.bar()
```

```python
# computing top states in terms of sales from first 10 observations
top_states_s = business.groupby("State").Sales.sum().nlargest(n=10)

# computing top states in terms of profit from first 10 observations
top_states_p = business.groupby("State").Profit.sum().nlargest(n = 10)

plt.style.use('seaborn')
top_states_s.plot(kind = 'bar',figsize = (10,5),fontsize = 14)
top_states_p.plot(kind = 'bar',figsize = (10,5),fontsize = 14, color = 'red')
plt.xlabel('States',fontsize = 15)
plt.ylabel('Total sales',fontsize = 15)
plt.title("Top 10 states Sales vs Profit",fontsize = 15)
plt.show()
```
